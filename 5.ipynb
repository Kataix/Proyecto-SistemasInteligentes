```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

#Se cargan los datos del archivo y se transforman en un Dataframe

d3 = np.load("Datos 3D/puntos_3d.npy")
df = pd.DataFrame(d3)
#le asignamos columnas x,y,z

df.columns = ["x","y","z"]

#se crea una figura en 3 dimensiones

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')

#se le coloca los datos del Dataframe a la figura

ax1.scatter(df['x'],df['y'],df['z'], c = 'r' , marker = 'o')

#Se crea otra figura en 3 dimensiones

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')

#se le coloca los datos del Dataframe a la figura

ax2.scatter(df['x'],df['y'],df['z'], c = 'r' , marker = 'o')

#centroide de color azul

ax2.scatter(df['x'].mean(),df['y'].mean(),df['z'].mean(), c = 'b' , marker = 'o', linewidths=9 )

plt.show()
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()

#Se separa la informacion del iris
x = iris.data[:,:2] #datos
y = iris.target #objetivo

#Se grafican los datos con matplotlib
plt.scatter(x[:,0],x[:,1], c = y, edgecolor = 'k')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')        
plt.show()
                     
```
